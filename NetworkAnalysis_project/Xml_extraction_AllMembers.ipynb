```python
#Main Code Chunk_part1
#create an empty dataframe 
members_df = pd.DataFrame()
#read the file containing names and ids of all the required government agencies
gov_agencies = pd.ExcelFile('C:/Users/Arko/Downloads/Studies/PersonalResearch/NetworkProject/Government_Agencies.xlsx')
gov_agencies = gov_agencies.parse('Sheet1')
#print(gov_agencies)
#gov_agencies['ID'].head()
#for each of the govr. agencies above, get details of the leaders
for k in gov_agencies.index:
    agencynumber = gov_agencies['ID'].ix[k]
    link = "http://api.littlesis.org/entity/"+ str(agencynumber) + "/related.xml?_key="
    key = "f08e0f2badde4762054a41cfa1fc60fba80d99a6"
    apilink = link+key
    #print(apilink)
    response = requests.get(apilink)
    #response code 200 indicates proper extraction
    #print(response.status_code)
    if response.status_code != 200:
        print(apilink)
        print("Problem while retrieving data from following link.Aborting.")
        break
    tree = ET.fromstring(response.content)
    #print(tree.getchildren())
    #get agency details
    agency_details = tree.findall('Data/Entity/')
    #print(len(agency_details))
    #get all the Related Members
    members = tree.findall('Data/RelatedEntities/Entity')
    #print(len(members))
    if len(members) == 0:
        print("Members not found")
        members_df = members_df.append({'AgencyId':agency_details[0].text,'AgencyName':agency_details[1].text,'Description':agency_details[2].text,
                                        'MemberId':'Not Found','MemberName':'Not Found','Position':'Not Found',
                                        'StartDate':'Not Found','EndDate':'Not Found','IsCurrentPosition':'Not Found',
                                        'RelationId':'Not Found','RelationDescp':'Not Found','RelationStartDate':'Not Found',
                                        'RelationEndDate':'Not Found','RelationIsCurrent':'NotFound','RelationAsBoardMember':'Not Found','RelationAsExecutive':'Not Found','RelationAsEmployee':'Not Found'},
                                        ignore_index=True)
    for i in range(len(members)):
        #agency_id should be same as agencynumber
        agency_id = agency_details[0].text
        agency_name = agency_details[1].text
        agency_desc = agency_details[2].text
        member_id = members[i][0].text
        member_name = members[i][1].text
        member_position = members[i][2].text
        member_startdate = members[i][4].text
        member_enddate = members[i][5].text
        #current position 1 indicates Yes
        member_ifcurrentposition = members[i][7].text
        #get the members or organizations to which the member is related
        #print(len(members[i][14]))
        for j in range(len(members[i][14])):
            related_id = members[i][14][j][0].text
            related_desc = members[i][14][j][4].text
            related_startdate = members[i][14][j][8].text
            related_enddate = members[i][14][j][9].text
            related_iscurrent = members[i][14][j][10].text
            related_isboard = members[i][14][j][15].text
            related_isexec = members[i][14][j][16].text
            related_isemployee = members[i][14][j][17].text
            #add to data frame
            members_df = members_df.append({'AgencyId':agency_details[0].text,'AgencyName':agency_details[1].text,
                                            'Description':agency_details[2].text,'MemberId':member_id,'MemberName':member_name,
                                            'Position':member_position,'StartDate':member_startdate,'EndDate':member_enddate,
                                            'IsCurrentPosition':member_ifcurrentposition,'RelationId':related_id,
                                            'RelationDescp':related_desc,'RelationStartDate':related_startdate,
                                            'RelationEndDate':related_enddate,'RelationIsCurrent':related_iscurrent,
                                            'RelationAsBoardMember':related_isboard,'RelationAsExecutive':related_isexec,
                                            'RelationAsEmployee':related_isemployee},ignore_index=True)
#rearrange the columns
members_df = members_df[['AgencyId','AgencyName','Description','MemberId','MemberName','Position','StartDate','EndDate',
                         'IsCurrentPosition','RelationId','RelationDescp','RelationStartDate','RelationEndDate',
                         'RelationIsCurrent','RelationAsBoardMember','RelationAsExecutive','RelationAsEmployee']]
#Note: Need to tackle the data formats
#write the final leaders dataframe to a csv file
members_df.to_csv("C:/Users/Arko/Downloads/Studies/PersonalResearch/NetworkProject/MonetaryAndFinancialAgencies_Members.csv")
```
