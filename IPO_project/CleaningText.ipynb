```python
import re, os, sys
import codecs
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords
from bs4 import BeautifulSoup
import string
```

```python
def clean_text(data):        
    # now remove the java script
    p = re.compile(r'< script[^<>]*?>.*?< / script >')
    data = p.sub('', data)
    # remove the css styles
    p = re.compile(r'< style[^<>]*?>.*?< / style >')
    data = p.sub('', data)
    # remove html comments
    p = re.compile(r'')
    data = p.sub('', data)
    # remove all the tags
    p = re.compile(r'<[^<]*?>')
    data = p.sub('', data)
    data = p.sub(' ',data)
    #remove stopwords
    cachedStopWords=stopwords.words('english')
    data = ' '.join([word for word in data.split() if word not in cachedStopWords])
    #remove punctuations
    data = "".join(l for l in data if l not in string.punctuation)
    # replace the series of '....' with a whitespace
    p = re.compile('&....;')
    # remove the extra whitespaces
    data = ' '.join(data.split())
    #convert entire text to lower
    data = data.lower()
    return(data)
```

```python
path = "C:/Users/Arko/Downloads/Studies/PersonalResearch/IPO_project/DATA/main/"
os.chdir(path)
dirs = os.listdir(path)
for file in dirs:
    with open (file,"r") as myfile:
        text1=myfile.read()
        #clean the text
        text2= clean_text(text1)
        #write the cleaned data to a text file
        fname = str(file).split("_")[0]
        print(fname)
        name="C:/Users/Arko/Downloads/Studies/PersonalResearch/IPO_project/DATA/CleanedData/"+fname+'_'+"clean"+".txt"
        file1 = open(name,'a')
        file1.write(text2)
        file1.close()
```
