```python
#read the file with fdate and odate information
df_dates = pd.read_excel("C:/Users/Arko/Downloads/Studies/PersonalResearch/IPO_project/DATA/s-1_filing_dates.xls")
df_dates = df_dates[pd.notnull(df_dates['cik'])]
#convert the datetime64 type to object for ease of comparison
df_dates['fdate'] = df_dates['fdate'].apply(lambda x: x.strftime('%Y-%m-%d')if not pd.isnull(x) else "Notdefined")
df_dates['odate'] = df_dates['odate'].apply(lambda x: x.strftime('%Y-%m-%d')if not pd.isnull(x) else "Notdefined")
#convert the cik number to str for future comparison
df_dates['cik'] = df_dates['cik'].astype(int).astype(str)
#df_dates['cik'].head()
```

```python
for file1 in dirs:
    li = []
    with open (file1,"r") as myfile1:
        #get name of the file1
        name1 = file1.split("_")[0].lstrip("0")
        #get fdate and odate of the file1. If fdate is empty store odate
        fdate1 = df_dates[df_dates["cik"] == name1].iat[0,6]
        if(fdate1 == 'Notdefined'):
            fdate1 = df_dates[df_dates["cik"] == name1].iat[0,1]
        text1 = myfile1.read().split()
        text1 = ' '.join(text1[1:])
        #print(text1)
        for doc in dirs:
            with open (doc,"r") as myfile1:
                name2 = doc.split("_")[0].lstrip("0")
                print(name2)
                 #get fdate and odate of the file1. If fdate is empty store odate
                fdate2 = df_dates[df_dates["cik"] == name2].iat[0,6]
                if(fdate2 == 'Notdefined'):
                    fdate2 = df_dates[df_dates["cik"] == name2].iat[0,1]
                #print("Reading file",myfile1)
                text2 = myfile1.read().split()
                text2 = ' '.join(text2[1:])
                if(fdate1 < fdate2):
                    similarity_ratio = SequenceMatcher(None,text1,text2).ratio()
                    #print(similarity_ratio)
                elif(fdate1 > fdate2):
                    similarity_ratio = SequenceMatcher(None,text2,text1).ratio()
                    #print(similarity_ratio)
                #if both fdates are equal,the cik created first gets preference
                elif(fdate1 == fdate2):
                    cik1 = int(name1)
                    #print(cik1)
                    cik2 = int(name2)
                    if(cik1 < cik2):
                        similarity_ratio = SequenceMatcher(None,text1,text2).ratio()
                    else:
                        similarity_ratio = SequenceMatcher(None,text2,text1).ratio()
                li.append(round(similarity_ratio,4))
                #print(li)
        df = df.append({'cik':file1[:-4], 'similarity':str(li)}, ignore_index=True)
#remove the empty columns
#print(df)
#print(df[pd.notnull(df['similarity'])])
df.to_csv('C:/Users/Arko/Downloads/Studies/PersonalResearch/IPO_project/similarity_anothermeasure.csv')
```
